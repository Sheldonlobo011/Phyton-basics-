<b>Q#1:</b> Consider the Contents.txt file. Using regular expressions,write a program to find: 
<ul>
    <li>Full email addresses.</li> 
    <li>words with exactly 8 letters using regex.</li>
    <li>The numbers starting with 212.</li>
    <li>The html tags that are more than 4 letters. Html tags can be found inside <> characters and closing html tags can be found in the same format after / character. </> </li>
<li>You are given stock prices for related financial tickers. (Symbols representing companies in the stock market).Find the tickers mentioned in the report. i.e.: TSLA, NFLX </li>

```python
# Write Your Answer for Q#1 Over Here
import re
#importing contents.txt file from local system
with open("Contents.txt","r") as f:
  data=f.read()

#printing the full email address 
print("Full Email Addresses is: ")
#for loop 
for j in re.findall(r"\S+@\S+",data):
  print(j)
#displaying the words with exactlly 8 letters 
print("Words with exactly 8 letters are: ")

for j in re.findall(r"\b[a-zA-Z]{8}\b",data):
  print(j)

#printing the output for numbers starting with 212 
print("The numbers starting with 212 are: ")
for j in re.findall(r'212[0-9-]*' , data):
  print(j)

#dispalying the output for tags <> with 4 letters 
print("The html tags that are more than 4 letters :")

for j in re.findall(r'<.*?>' , data):
  print(j)
```

```python
# Write your answer here (part1)
#importing pandas 
import pandas as pd
#reading the data set black friday
data = pd.read_csv("BlackFriday.csv")
#describing the data set
data.describe()
```

```python
# Write your answer here (part2)
gender_set = {'User_ID': list(data['User_ID']), 'Gender': list(data['Gender']), 'Age': list(data['Age']) } #creation of dataset with columns UserID, Gender and age
customer_data = pd.DataFrame(gender_set)
#removing the duplicate rows from the table 
customer_data = customer_data.drop_duplicates() 
#printing the new dataset
print(customer_data)
#printing the males and females count using groupby method
GCount = customer_data.groupby('Gender')['Gender'].count() 
print(GCount)
```

```python
# Write your answer here (part3)
maxcityage_data = data[['City_Category', 'Age', 'Purchase']].groupby(['Age', 'City_Category']).max()
maxcityage_data = maxcityage_data.sort_values(['Purchase'], ascending=False).reset_index()
print(maxcityage_data)
```

```python
# Write your answer here (part4)
mingendermarital_data = data[['Gender', 'Marital_Status', 'Purchase']].groupby(['Gender', 'Marital_Status']).min()
mingendermarital_data = mingendermarital_data.sort_values(['Purchase'], ascending=True).reset_index()
mingendermarital_data
```
