<b>Q#3</b> Consider the <b><i>Team</i></b> file. This file contains a winning teams. The first line in the file is the name of the team that won in 1903, and the last line is the name of the team that won in 2006. Write a program that reads this file and creates a dictionary in which the keys are the names of the teams, and each key’s associated value is the number of times the team has won. The program should also create a dictionary in which the keys are the years, and each key’s associated value is the name of the team that won that year.
The program should prompt the user for a year in the range of 1903 through 2006. It should then display the name of the team that won that year, and the number of times that team has won.

```python
# Write Your Answer for Q#3 Over Here

# setting the starting year of the data in the dictionary
starting_year = 1903

# Dictionary in which value is the number of times the team has won
winDict = {}
year_TeamDict = {}

# Open file from local storange for reading 
with open('Team.txt', 'r') as file:
    
    # Loop of each line in in the local file
    for line in file:
        
        line = line.strip()
        
# searching team win count 
        if line in winDict:
            winDict[line] += 1
# If not found    
        else:
            
            winDict[line] = 1
        year_TeamDict[starting_year] = line 
        starting_year += 1 

# Taking input from the user about the year of thier choice    
year_Search = int(input('Enter a year in the range 1903-2006: '))

# Print the team name asked by the user
print('The team that won in ', year_Search, ' is the ', year_TeamDict[year_Search], '.', sep = '')

# Printing the total no of championship years
print('They won ', winDict[year_TeamDict[year_Search]], ' times.', sep = '')
```

Q#4: Write a program that prompts the user to enter information (model, color, make, and year) about as many car as he/she wishes. Each time the user enters information about a car, the information is stored in a dictionary, then the dictionary is pickled and saved to a file named car_info.dat. Each time the program starts, it should retrieve the dictionary from the file and unpickle it.

```python
# Write Your Answer for Q#4 Over Here
#Importing pickle
import pickle

# Create an empty dictionary called car which will store the vlaue of the car 
car={}
i=1 

# Opening the file to write into it
car_data=open("car_info_file.dat","wb")
again='y'
while again=='y':
    
# Prompting inputs from the user about the model, make and year of manufacture of the car    
        model=input("Enter the model of the car:")
        make=input("Enter the make of the car:")
        year=input("Enter the year the car is manufactured:")
        color=input('Enter the color of the car')
        car[i]=[model,make,year,color]
        i=i+1
# Asking user whether we he want to see more information      
        again=input("Do you want to another car detials Yes/No- y/n?: ")
    
# Using pickle function to load dictionary into file    
pickle.dump(car,car_data)
car_data.close()

# Opening car info file in read mode 
car_data_open=open("car_info_file.dat","rb") 

# Unpickling file to view the content
file_inf=pickle.load(car_data_open)
car_data_open.close()
print("added to car_info file:")
# Printing the dictionary
print(file_inf) 
```
